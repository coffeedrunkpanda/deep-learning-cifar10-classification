# Data Exploration

The dataset we are going to use is [cifar10](https://keras.io/api/datasets/cifar10/), a dataset of 50,000 32x32 color training images and 10,000 test images, labeled over 10 categories.

The classes: 


| Label | Description |
|-------|-------------|
| 0     | airplane    |
| 1     | automobile  |
| 2     | bird        |
| 3     | cat         |
| 4     | deer        |
| 5     | dog         |
| 6     | frog        |
| 7     | horse       |
| 8     | ship        |
| 9     | truck       |


```python
from tensorflow.keras.datasets import cifar10
(X_train, y_train), (X_test, y_test) = cifar10.load_data()
```

```python
# Dataset integraty check

assert X_train.shape == (50000, 32, 32, 3)
assert X_test.shape == (10000, 32, 32, 3)
assert y_train.shape == (50000, 1)
assert y_test.shape == (10000, 1)
```

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 
               'dog', 'frog', 'horse', 'ship', 'truck']
```

## Visualizing samples

```python
import matplotlib.pyplot as plt

def plot_sample(X, y, class_names):

    plt.figure(figsize=(10, 5))
    for i in range(10):
        plt.subplot(2, 5, i+1)
        plt.imshow(X[i])
        plt.title(class_names[y[i][0]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

```

### Train samples (First 10)

```python
plot_sample(X_train, y_train, class_names)
```

### Test samples (First 10)

```python
plot_sample(X_test, y_test, class_names)
```

## Samples per class

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()

def plot_class_freq(y, class_names, title):
    label_counts = np.unique_counts(y)

    # values = label_counts.values
    counts = label_counts.counts

    sns.barplot(x =class_names, y = counts)
    plt.xticks(rotation=45)
    plt.xlabel('Class')
    plt.ylabel('N Samples')
    plt.title(title)

    plt.show()

```

```python
plot_class_freq(y_train, class_names, title="Samples per Class: Training Set")
```

```python
plot_class_freq(y_train, class_names, title="Samples per Class: Testing Set")
```
